```python
import torch
from matplotlib import  pyplot as plt
import numpy as np
import cv2
```

```python
model = torch.hub.load('ultralytics/yolov5','yolov5s')
```

```python
img = "i.png"
results = model(img)
results.print()
```

```python
%matplotlib inline
plt.imshow(np.squeeze(results.render()))
plt.show()
```

```python
cap = cv2.VideoCapture("lev.mp4")
while cap.isOpened():
    ret, frame = cap.read()
    cv2.imshow('Yolo',frame)

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
```
